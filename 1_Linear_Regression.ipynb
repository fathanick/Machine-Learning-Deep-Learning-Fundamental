```
X = data.iloc[:,0]
Y = data.iloc[:,1]
```

```
#sedikit atur parameter plot
plt.rcParams['figure.figsize'] = (12.0, 10.0)
```

```
plt.scatter(X,Y)
plt.show()
```

```
a = 0
b = 0
L = 0.000001 #learning rate
iteration = 25 #jumlah iterasi
n = float(len(X)) #banyaknya data pada X
for i in range(iteration):
  Y_pred = a*X + b #nilai prediksi Y untuk X yang diberikan

  Error = 1/2*sum((a*X + b - Y)**2) # slide Pak Ridho

  D_a = sum( X * (a*X + b - Y)) # Partial derivative a
  D_b = sum( (a*X + b - Y)) # Partial derivative b
  a = a - L * D_a
  b = b - L * D_b
  print ("Iterasi ke: ", i+1, "a: ", a, "b: ", b, "Error: ", Error)
  plt.scatter(X,Y)
  plt.plot([min(X),max(X)],[min(Y_pred),max(Y_pred)], color='red')
```

```
print('Regression Model: Y= {}' .format(float(a)),'X + {}' .format(float(b)))
```

###Linear Regression with Scikit-Learn

```
from sklearn.linear_model import LinearRegression
```

```
X2 = data.iloc[:,0].values.reshape(-1,1)
Y2 = data.iloc[:,1].values
```

```
model = LinearRegression()
model.fit(X2,Y2)
```

```
a2 = model.coef_
b2 = model.intercept_
Y2_pred = a2*X2 + b2
print('Regression Model: Y= {}' .format(float(a2)),'X + {}' .format(float(b2)))
```

```
plt.scatter(X2, Y2)
plt.plot([min(X2), max(X2)], [min(Y2_pred), max(Y2_pred)], color='red')
```
